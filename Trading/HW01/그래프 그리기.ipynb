y = ${2x^2 +3x +8}$


y = $3x - 11$

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

import numpy as np
```

```python
x_array = np.linspace(-5,5,50)
y_1 = np.array([2 * np.power(x,2) + 3*x + 8 for x in x_array])
y_2 = 3* x_array - 11

plt.figure(figsize = (15,6))
plt.subplot(1,2,1)
plt.plot(x_array, y_1, label = '$2x^2 +3x +8$')
plt.legend(loc = 'best', fontsize = 15)


plt.subplot(1,2,2)
plt.plot(x_array, y_2, label = '$3x - 11$')
plt.legend(loc = 'best', fontsize = 15)
```
