# 로그인은 수동으로

# 게시판 메뉴 클릭하여 크롤링

```python
def click_page(css):
    import time
    
    # iframe으로 전환
    try:
        driver.switch_to_frame('cafe_main')
    except:
        pass
    
    driver.get("https://cafe.naver.com/anycallusershow")
    driver.find_element_by_css_selector(css).click()
    
    time.sleep(3)
```

```python
def crawl_text(text_ls, datgle_ls):
    # 게시글에 들어온 다음, 크롤링을 진행한다
    try:
        # 게시글 크롤링
        txt = driver.find_element_by_css_selector('#tbody').text.replace('\n', ' ')
        text_ls.append(txt)

        # 댓글 크롤링
        for j in range(1,50):
            try:                   
                datgle = driver.find_element_by_css_selector('#cmt_list > li:nth-child(%s) > div > p > span.comm_body'%j).text.replace('\n', ' ')

                datgle_ls.append(datgle)

            except:
                #댓글이 아예 없는 경우에는 그냥 나간다
                if j == 1 : 
                    drive.back()
                    break
                pass
        
        drive.back()

    # 크롤링이 안되면 뒤로 돌아간다.
    except:
        driver.back()
        pass
    
    
    return text_ls, datgle_ls
    
```

```python
def crawl_one_page(text_ls, datgle_ls):
    import time
    time.sleep(1)
    
    board_type = 1
    
    # iframe으로 전환
    try:
        driver.switch_to_frame('cafe_main')
    except:
        pass
    
    
    # 1번 타입 게시판에서 크롤링
    if board_type == 1:
    
        # 개별 게시글에 접근
        for i in range(1,11):
            try:
                driver.switch_to_default_content()
                driver.execute_script("window.scrollTo(0, %s)"%(150*i)) 
                time.sleep(0.5)

                driver.switch_to_frame('cafe_main')

                # 게시글에 접근
                driver.find_element_by_css_selector('#main-area > ul.article-movie-sub > li:nth-child(%s) > dl > dt > a.m-tcol-c'%(i)).click()
                
                # 게시글과 댓글 크롤링
                text_ls, datgle_ls = crawl_text(text_ls, datgle_ls)
                
            except : 
                board_type = 2
                break
    
    # 마지막 게시글에 접근한 후에는 스크롤을 내려서 다음 페이지로 넘어갈 준비를 한다.
    if board_type == 1:
        driver.switch_to_default_content()
        driver.execute_script("window.scrollTo(0, %s)"%(150*i)) 
        time.sleep(0.3)

        driver.switch_to_frame('cafe_main')
                       
            
            
    # 2번 타입 게시판에 적용되는 포맷
    if board_type == 2:
        
        for i in range(15):
            try: driver.switch_to_frame('cafe_main')
            except: pass
            
            try:
                # 스크롤 내리기
                driver.switch_to_default_content()
                driver.execute_script("window.scrollTo(0, %s)"%(50*i)) 
                time.sleep(0.3)
                driver.switch_to_frame('cafe_main')
                
                # 게시글에 접근
                driver.find_element_by_css_selector('#main-area > div:nth-child(7) > form > table > tbody > tr:nth-child(%s) > td.board-list > span > span > a.m-tcol-c'%(2*i +1)).click()
                
                # 게시글과 댓글 크롤링
                text_ls, datgle_ls = crawl_text(text_ls, datgle_ls)
                
                
                # 페이지 크롤링을 마친 후, 다음 페이지로 넘어가기 위해 스크롤을 내려둔다
                if i == 14:
                    # 스크롤 내리기
                    driver.switch_to_default_content()
                    driver.execute_script("window.scrollTo(0, %s)"%(50*i)) 
                    time.sleep(0.3)
                    driver.switch_to_frame('cafe_main')
                    
            except:
                return text_ls, datgle_ls

 
        
    return text_ls, datgle_ls
```

```python
def crawl_multiple_pages(n_page):
    import time 
    
    text_ls = []
    datgle_ls = []
    
    try:   driver.switch_to_frame('cafe_main')
    except:   pass

    for cnt in range(n_page):
        
        if cnt == 0:
            for k in range(2,12):
                # 게시글과 댓글 크롤링
                text_ls,datgle_ls = crawl_one_page(text_ls, datgle_ls)

                try:
                    # 페이지를 이동한다
                    driver.find_element_by_css_selector('#main-area > div.prev-next > table > tbody > tr > td:nth-child(%s) > a'%(k)).click()

                #안되면 그만한다
                except:
                    return text_ls, datgle_ls
                    
            cnt += 1

        else:
            for k in range(3,13):
                text_ls, datgle_ls = crawl_one_page(text_ls, datgle_ls)
                
                try:
                    driver.find_element_by_css_selector('#main-area > div.prev-next > table > tbody > tr > td:nth-child(%s) > a'%(k)).click()
                except:  return text_ls, datgle_ls
               
```

```python
crawl_dict = {#'NOTE_9' : '#menuLink924',
              #'S9'  : '#menuLink914',
              #'NOTE_8' : '#menuLink902',
              #'S8' : '#menuLink877',
              'NOTE_FE' : '#menuLink891',
              #'S7' : '#menuLink833',
              #'NOTE_5' : '#menuLink809',
              #'S6' : '#menuLink789',
              #'A_Series' : '#menuLink769',
              #'J_Series' : '#menuLink845',
             }
```

```python
text_result_df = pd.DataFrame()
datgle_result_df = pd.DataFrame()

for key,value in crawl_dict.items():
    
    # 해당 게시판으로 이동
    #click_page(value)
    
    # iframe으로 전환
    try: driver.switch_to_frame('cafe_main')
    except: pass
    
    text_ls, datgle_ls = crawl_multiple_pages(n_page = 20)
    
    text_df = pd.DataFrame(text_ls, columns = [key + '_게시글'])
    datgle_df = pd.DataFrame(datgle_ls, columns = [key + '_댓글'])
    
    text_df.to_csv('Data/Naver_Cafe/Naver_Cafe_%s_text_%s.csv'%(key, len(text_df)), index = False)
    datgle_df.to_csv('Data/Naver_Cafe/Naver_Cafe_%s_datgle_%s.csv'%(key,len(datgle_df)), index = False)
    
    text_result_df = pd.concat([text_result_df, text_df], axis = 1, copy = False)
    datgle_result_df = pd.concat([datgle_result_df, datgle_df], axis = 1, copy = False)
    
    
```
