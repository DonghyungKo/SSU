```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
```

```python
path = "D:\chromedriver.exe"
driver = webdriver.Chrome(path)
driver.get("https://www.cetizen.com/")
driver.implicitly_wait(1)
```

```python
def crawl_text(text_ls):
    
    # 한 페이지당 게시물 10개
    for i in range(10):
        try:
            driver.find_element_by_css_selector('#review_list > div:nth-child(1) > div:nth-child(%s) > div:nth-child(2) > div:nth-child(1) > a'%(2*i + 1)).click()
            text = driver.find_element_by_css_selector('body > div:nth-child(14) > div:nth-child(3) > div:nth-child(5) > div > pre').text
            
            text = text.replace('\n', ' ')
            text_ls.append(text)
        
        except:
            try: 
                text = driver.find_element_by_css_selector('body > div:nth-child(15) > div:nth-child(3) > div:nth-child(5) > div > pre').text
                
                text = text.replace('\n', ' ')
                text_ls.append(text)
        
            except: driver.back()
            
        
        driver.back()

    return text_ls
```

```python
driver.find_element_by_css_selector('body > div:nth-child(15) > div:nth-child(3) > div:nth-child(5) > div > pre').text
```

```python
def crawl_junmoon_review(key_word, n_page):
    
    text_ls = []
    
    driver.find_element_by_css_selector('#menu_2 > a').click()
    driver.find_element_by_css_selector('#r_menu > ul:nth-child(2) > li:nth-child(1) > a > span').click()
    
    driver.find_element_by_css_selector('#keyword_p').click()
    driver.find_element_by_css_selector('#keyword_p').send_keys(key_word)
    driver.find_element_by_css_selector('#ReviewSearchForm > div:nth-child(7) > div > div:nth-child(2) > input').click()
    
    for cnt in range(n_page):

        if cnt == 0 :
            for i in range(2, 12):
                text_ls = crawl_text(text_ls)

                #페이지 이동 
                try:
                    driver.find_element_by_css_selector('#page_no > div > span:nth-child(%s)'%i).click()
                except:
                    return text_ls

        else:
            for i in range(3, 13):
                text_ls = crawl_text(text_ls)

                # 페이지 이동
                try:
                    driver.find_element_by_css_selector('#page_no > div > span:nth-child(%s)'%i).click()
                except:
                    return text_ls

    return text_ls

    
```

```python
for key_word in ['삼성', '아이폰', 'LG']:
    text_ls = crawl_junmoon_review(key_word, 10)
    
    text_df = pd.DataFrame(text_ls, columns = [key_word])
    text_df.to_csv('cetizen_review_%s.csv'%(key_word.replace('삼성','Samsung').replace('아이폰','IPhone')))
```
